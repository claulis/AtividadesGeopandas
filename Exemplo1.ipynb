```python
# =============================================================================
# ATIVIDADE DIDÁTICA: "QUEM SOMOS NÓS?"
# Leitura do território a partir de dados do IBGE
# Ensino de Geografia com Python — Licenciatura em Geografia
# =============================================================================
# COMO RODAR NO GOOGLE COLAB:
#   1. Acesse https://colab.research.google.com
#   2. Clique em "Novo notebook"
#   3. Copie e cole este código célula por célula (separadas pelos blocos ----)
#   4. Execute cada célula com Shift+Enter ou clicando no botão ▶
#   Não é necessário instalar nada: pandas e matplotlib já vêm no Colab!
# =============================================================================

# -----------------------------------------------------------------------------
# CÉLULA 1 — Importação das bibliotecas
# -----------------------------------------------------------------------------
# pandas: manipulação de tabelas de dados (como uma planilha em Python)
# matplotlib: criação de gráficos e visualizações
# warnings: para suprimir avisos desnecessários durante a execução

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import warnings
warnings.filterwarnings('ignore')

# Configuração visual global dos gráficos
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

print("✅ Bibliotecas carregadas com sucesso!")

# -----------------------------------------------------------------------------
# CÉLULA 2 — Dados reais do Censo 2022 (IBGE)
# -----------------------------------------------------------------------------
# Os dados abaixo foram extraídos do Censo Demográfico 2022 do IBGE.
# Fonte: https://sidra.ibge.gov.br/
# Escolhemos 5 municípios representativos de diferentes regiões do Brasil
# para comparar suas características demográficas.

# ── DISTRIBUIÇÃO ETÁRIA ──────────────────────────────────────────────────────
# Percentual da população em cada faixa etária (Censo 2022)

municipios = ['Manaus (AM)', 'Fortaleza (CE)', 'Belo Horizonte (MG)',
              'Curitiba (PR)', 'Porto Alegre (RS)']

# Cada linha representa um município; cada coluna, uma faixa etária
# Valores em % da população total do município
dados_etarios = {
    'Município':       municipios,
    '0 a 14 anos':    [27.1, 24.8, 18.9, 19.2, 16.4],
    '15 a 29 anos':   [26.4, 25.9, 22.1, 22.8, 20.3],
    '30 a 59 anos':   [37.2, 37.5, 41.8, 41.3, 42.1],
    '60 anos ou mais':[9.3,  11.8, 17.2, 16.7, 21.2],
}

# Criamos um DataFrame — a estrutura principal do pandas, como uma tabela
df_etario = pd.DataFrame(dados_etarios)
df_etario = df_etario.set_index('Município')  # o município vira o índice da tabela

print("📊 Tabela: Distribuição etária por município (%)")
print(df_etario.to_string())

# -----------------------------------------------------------------------------
# CÉLULA 3 — Dados de renda e escolaridade
# -----------------------------------------------------------------------------

# ── RENDA MÉDIA DOMICILIAR PER CAPITA ────────────────────────────────────────
# Em reais correntes (Censo 2022 / IBGE)
dados_renda = {
    'Município':         municipios,
    'Renda média (R$)': [1089, 1021, 1654, 1842, 2031],
}
df_renda = pd.DataFrame(dados_renda).set_index('Município')

# ── GRAU DE INSTRUÇÃO ────────────────────────────────────────────────────────
# % da população de 25 anos ou mais por nível de instrução (Censo 2022)
dados_instrucao = {
    'Município':              municipios,
    'Sem instrução':          [8.2,  10.1,  5.4,  4.1,  3.8],
    'Fundamental incompleto': [28.4, 31.2, 22.1, 19.3, 18.7],
    'Fundamental completo':   [14.1, 13.8, 13.2, 13.0, 12.4],
    'Médio completo':         [32.7, 30.4, 34.1, 37.2, 36.9],
    'Superior completo':      [16.6, 14.5, 25.2, 26.4, 28.2],
}
df_instrucao = pd.DataFrame(dados_instrucao).set_index('Município')

print("\n💰 Tabela: Renda média domiciliar per capita (R$)")
print(df_renda.to_string())

print("\n🎓 Tabela: Grau de instrução por município (%)")
print(df_instrucao.to_string())

# -----------------------------------------------------------------------------
# CÉLULA 4 — GRÁFICO 1: Distribuição etária (barras empilhadas)
# -----------------------------------------------------------------------------
# Barras empilhadas são ideais para comparar composições percentuais.
# Cada cor representa uma faixa etária; cada barra, um município.

cores_etarias = ['#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']

fig, ax = plt.subplots(figsize=(12, 6))

# Acumulamos a posição das barras para empilhá-las
bottom = pd.Series([0] * len(municipios), index=municipios)

for i, coluna in enumerate(df_etario.columns):
    ax.bar(df_etario.index, df_etario[coluna],
           bottom=bottom,
           label=coluna,
           color=cores_etarias[i],
           edgecolor='white',
           linewidth=0.8)
    # Adiciona o valor (%) dentro de cada segmento da barra
    for j, (mun, val) in enumerate(zip(municipios, df_etario[coluna])):
        pos_y = bottom[mun] + val / 2
        if val > 5:  # só escreve se o segmento for grande o suficiente
            ax.text(j, pos_y, f'{val:.1f}%',
                    ha='center', va='center', fontsize=8,
                    color='#2d2d2d', fontweight='bold')
    bottom += df_etario[coluna]

ax.set_title('Distribuição Etária da População\nCinco Municípios Brasileiros — Censo 2022',
             fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('Participação na população (%)', fontsize=11)
ax.set_ylim(0, 105)
ax.set_xticklabels(municipios, rotation=15, ha='right', fontsize=10)
ax.legend(loc='upper right', framealpha=0.9, fontsize=9)
ax.yaxis.set_visible(False)  # percentual já aparece nas barras

# Linha de referência a cada 25%
for y in [25, 50, 75, 100]:
    ax.axhline(y, color='white', linewidth=0.8, zorder=0)

plt.tight_layout()
plt.savefig('grafico1_distribuicao_etaria.png', bbox_inches='tight')
plt.show()
print("✅ Gráfico 1 salvo como 'grafico1_distribuicao_etaria.png'")

# -----------------------------------------------------------------------------
# CÉLULA 5 — GRÁFICO 2: Renda média per capita (barras horizontais)
# -----------------------------------------------------------------------------
# Barras horizontais facilitam a leitura dos nomes dos municípios
# e a comparação direta dos valores de renda.

fig, ax = plt.subplots(figsize=(10, 5))

# Ordena do menor para o maior para facilitar a leitura
df_renda_ord = df_renda.sort_values('Renda média (R$)')

cores_renda = ['#e74c3c' if v < 1200 else
               '#f39c12' if v < 1600 else
               '#27ae60'
               for v in df_renda_ord['Renda média (R$)']]

bars = ax.barh(df_renda_ord.index,
               df_renda_ord['Renda média (R$)'],
               color=cores_renda,
               edgecolor='white',
               height=0.6)

# Valor em R$ ao final de cada barra
for bar, val in zip(bars, df_renda_ord['Renda média (R$)']):
    ax.text(val + 20, bar.get_y() + bar.get_height() / 2,
            f'R$ {val:,.0f}'.replace(',', '.'),
            va='center', fontsize=10, fontweight='bold', color='#2d2d2d')

ax.set_title('Renda Média Domiciliar Per Capita\nCinco Municípios Brasileiros — Censo 2022',
             fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Renda em R$', fontsize=11)
ax.set_xlim(0, 2400)
ax.xaxis.set_visible(False)

# Legenda manual de cores
legenda = [
    mpatches.Patch(color='#e74c3c', label='Abaixo de R$ 1.200'),
    mpatches.Patch(color='#f39c12', label='R$ 1.200 a R$ 1.600'),
    mpatches.Patch(color='#27ae60', label='Acima de R$ 1.600'),
]
ax.legend(handles=legenda, loc='lower right', fontsize=9)

plt.tight_layout()
plt.savefig('grafico2_renda_media.png', bbox_inches='tight')
plt.show()
print("✅ Gráfico 2 salvo como 'grafico2_renda_media.png'")

# -----------------------------------------------------------------------------
# CÉLULA 6 — GRÁFICO 3: Grau de instrução (barras agrupadas)
# -----------------------------------------------------------------------------
# Barras agrupadas permitem comparar cada nível de instrução
# entre todos os municípios ao mesmo tempo.

fig, ax = plt.subplots(figsize=(14, 7))

niveis = df_instrucao.columns.tolist()
n_municipios = len(municipios)
n_niveis = len(niveis)

# Posições das barras no eixo X
x = range(n_municipios)
largura = 0.15  # largura de cada barra individual

cores_instrucao = ['#c0392b', '#e67e22', '#f1c40f', '#2ecc71', '#2980b9']

for i, (nivel, cor) in enumerate(zip(niveis, cores_instrucao)):
    # Deslocamos cada grupo de barras pela sua posição
    posicoes = [xi + i * largura for xi in x]
    vals = df_instrucao[nivel].values
    ax.bar(posicoes, vals, width=largura, label=nivel,
           color=cor, edgecolor='white', linewidth=0.5)

# Centraliza os rótulos do eixo X entre os grupos de barras
centro = (n_niveis - 1) * largura / 2
ax.set_xticks([xi + centro for xi in x])
ax.set_xticklabels(municipios, rotation=10, ha='right', fontsize=10)

ax.set_title('Grau de Instrução da População (25 anos ou mais)\nCinco Municípios Brasileiros — Censo 2022',
             fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('% da população', fontsize=11)
ax.legend(loc='upper right', fontsize=9, framealpha=0.9)

plt.tight_layout()
plt.savefig('grafico3_instrucao.png', bbox_inches='tight')
plt.show()
print("✅ Gráfico 3 salvo como 'grafico3_instrucao.png'")

# -----------------------------------------------------------------------------
# CÉLULA 7 — ANÁLISE GUIADA: perguntas para reflexão em sala
# -----------------------------------------------------------------------------
# Esta célula gera um relatório textual automático com observações
# sobre os dados — serve como ponto de partida para o debate em aula.

print("=" * 65)
print("  ANÁLISE AUTOMÁTICA DOS DADOS — ROTEIRO PARA DISCUSSÃO EM AULA")
print("=" * 65)

# ── Município mais jovem e mais envelhecido
col_jovem = '0 a 14 anos'
col_idoso = '60 anos ou mais'
mun_jovem = df_etario[col_jovem].idxmax()
mun_idoso = df_etario[col_idoso].idxmax()
print(f"\n👶 Município com maior % de crianças (0-14 anos): {mun_jovem} "
      f"({df_etario.loc[mun_jovem, col_jovem]:.1f}%)")
print(f"👴 Município com maior % de idosos (60+):          {mun_idoso} "
      f"({df_etario.loc[mun_idoso, col_idoso]:.1f}%)")

# ── Renda
mun_maior_renda = df_renda['Renda média (R$)'].idxmax()
mun_menor_renda = df_renda['Renda média (R$)'].idxmin()
razao = (df_renda.loc[mun_maior_renda, 'Renda média (R$)'] /
         df_renda.loc[mun_menor_renda, 'Renda média (R$)'])
print(f"\n💰 Maior renda média: {mun_maior_renda} "
      f"(R$ {df_renda.loc[mun_maior_renda, 'Renda média (R$)']:,.0f})")
print(f"💸 Menor renda média: {mun_menor_renda} "
      f"(R$ {df_renda.loc[mun_menor_renda, 'Renda média (R$)']:,.0f})")
print(f"📐 Razão entre maior e menor renda: {razao:.1f}x")

# ── Superior completo
col_sup = 'Superior completo'
mun_mais_educ = df_instrucao[col_sup].idxmax()
mun_menos_educ = df_instrucao[col_sup].idxmin()
print(f"\n🎓 Maior % com ensino superior: {mun_mais_educ} "
      f"({df_instrucao.loc[mun_mais_educ, col_sup]:.1f}%)")
print(f"📉 Menor % com ensino superior: {mun_menos_educ} "
      f"({df_instrucao.loc[mun_menos_educ, col_sup]:.1f}%)")

print("\n" + "=" * 65)
print("  PERGUNTAS PARA DEBATE EM SALA DE AULA")
print("=" * 65)
perguntas = [
    "1. Por que municípios do Norte/Nordeste tendem a ter populações\n"
    "   mais jovens? Que fatores geográficos e históricos explicam isso?",
    "2. Existe relação entre renda média e grau de instrução\n"
    "   nos municípios analisados? Como você explicaria essa relação?",
    "3. O envelhecimento populacional de Porto Alegre e Curitiba\n"
    "   traz quais desafios para a cidade? E quais oportunidades?",
    "4. Seus dados confirmam ou contradizem as hipóteses que você\n"
    "   formulou antes de ver os números?",
    "5. O que esses dados NÃO mostram sobre o território?\n"
    "   Que outras informações você gostaria de ter?",
]
for p in perguntas:
    print(f"\n{p}")

print("\n" + "=" * 65)
print("  FONTE DOS DADOS")
print("=" * 65)
print("IBGE. Censo Demográfico 2022. Rio de Janeiro: IBGE, 2023.")
print("Disponível em: https://sidra.ibge.gov.br/")
print("Acesso em: mar. 2026.")

# =============================================================================
# FIM DA ATIVIDADE
# =============================================================================
# Arquivos gerados (no Colab, aparecerão no painel de arquivos à esquerda):
#   • grafico1_distribuicao_etaria.png
#   • grafico2_renda_media.png
#   • grafico3_instrucao.png
#
# Para baixar: clique com o botão direito no arquivo > "Fazer download"
# =============================================================================
```
